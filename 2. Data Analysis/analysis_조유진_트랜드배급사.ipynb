## 독립/예술영화 트랜드를 잘 따라가는 배급사 찾기

해마다 대박을 친 영화의 장르가 다를까?
* 트랜드를 잘 따라간다는 기준 (조작적 정의) <br>
트렌드 분석 -> 수상작 -> 데이터 잘려서 볼 수 없음 <br>
좌석점유율은 사람들이 그만큼 봤다는 것을 의미한다<br>
극장에 와서 봤다는 것을 의미. <br>
이것은 선호하니까 와서 본것 <br>
<br>
1. 트렌드를 봐야하기 때문에 년도별로 흐름이 궁금하다 <br>
2. 장르별 좌석 점유율 -> 배급사별 장르 수를 세서  <br>
3. 일반 영화에서는 액션이 가장 높은데, 독립영화는 드라마가 가장 높다 <br>
4. 일반영화/독립영화 소비자들의 선호 장르가 다르다
<br><br>
스크린 수와 좌석점유율의 관계를 왜 찾아야하는가..<br>
http://star.ohmynews.com/NWS_Web/OhmyStar/at_pg.aspx?CNTN_CD=A0002194253

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family='AppleGothic')
kobis_tot = pd.read_csv("https://raw.githubusercontent.com/yujin-jo/mini-project-team1/main/1.%20Crawling/data/KOBIS_TOT.csv")

art_m = kobis_tot[(kobis_tot.개봉연도 >= 2014)]
연도별_전체_관객수 = art_m.groupby(['개봉연도'])['전국관객수'].sum().reset_index(name='개봉연도별_관객수')
연도별_장르_관객수 = art_m.groupby(['개봉연도','장르'])['전국관객수'].sum().reset_index()

# 연도별_장르_관객수.update(연도별_전체_관객수)
연도별_머지 = 연도별_장르_관객수.merge(연도별_전체_관객수, how='left')
연도별_머지
```

```python
# 좌석점유율 구하기
연도별_머지['좌석점유율'] = 연도별_머지['전국관객수'] / 연도별_머지['개봉연도별_관객수'] * 100
연도별_머지_top = 연도별_머지.sort_values(by="좌석점유율", ascending=False).groupby(["개봉연도"]).head(1)
연도별_머지_top.sort_values(by="개봉연도")

연도별_머지_top3 = 연도별_머지.sort_values(by="좌석점유율", ascending=False).groupby(["개봉연도"]).head(3)
연도별_머지_top3.sort_values(by=["개봉연도",'좌석점유율'], ascending=[True, False])
# 이렇게..독립영화의 장르가..편중되어있다....
```

장르별 좌석 점유율 확인 -> 배급사별 장르 수를 세기<br>
트렌드를 따르는 배급사 = 높은 좌석점유율을 따르는 배급사 <br>
드라마 장르를 내는 배급사????????????가 트렌드를 따르는 배급사라면..... <br>

```python
연도별_CGV제외_관객수 = art_m[(art_m.배급사.str.contains('CGV') == False)].groupby(['개봉연도'])['전국관객수'].sum().reset_index(name='개봉연도별_관객수')
연도별_CGV제외_장르_관객수 = art_m[(art_m.배급사.str.contains('CGV') == False)].groupby(['개봉연도','장르'])['전국관객수'].sum().reset_index()

연도별_CGV제외_머지 = 연도별_CGV제외_장르_관객수.merge(연도별_CGV제외_관객수, how='left')
연도별_CGV제외_머지

# 좌석점유율 구하기
연도별_CGV제외_머지['좌석점유율'] = 연도별_CGV제외_머지['전국관객수'] / 연도별_CGV제외_머지['개봉연도별_관객수'] * 100
연도별_CGV제외_머지_top = 연도별_CGV제외_머지.sort_values(by="좌석점유율", ascending=False).groupby(["개봉연도"]).head(1)
연도별_CGV제외_머지_top.sort_values(by="개봉연도")

# 연도별_머지_top3 = 연도별_머지.sort_values(by="좌석점유율", ascending=False).groupby(["개봉연도"]).head(3)
# 연도별_머지_top3.sort_values(by=["개봉연도",'좌석점유율'], ascending=[True, False])
# 이렇게..독립영화의 장르가..편중되어있다....
```
