```python
import pandas as pd
```

```python
naver = pd.read_csv('https://raw.githubusercontent.com/yujin-jo/mini-project-team1/main/1.%20Crawling/data/naver.csv')
cine21 = pd.read_csv('https://raw.githubusercontent.com/yujin-jo/mini-project-team1/main/1.%20Crawling/data/cine21_score.csv')
watcha = pd.read_csv('https://raw.githubusercontent.com/yujin-jo/mini-project-team1/main/1.%20Crawling/data/watcha_movie_score.csv')
kobis = pd.read_csv('https://raw.githubusercontent.com/yujin-jo/mini-project-team1/main/1.%20Crawling/data/kobis_tot.csv')
```

```python
kobis_2 = kobis[['순번','개봉연도','영화명','감독','등급','장르','전국매출액','전국관객수','전국스크린수','국적']]
```

```python
movie_merge = pd.merge(naver, cine21, how='left', on = '순번')
movie_merge_2 = pd.merge(movie_merge, watcha, how='left', on = '순번')
movie_merge_3 = pd.merge(movie_merge_2, kobis_2, how='left', on = '순번')
movie_fin = movie_merge_3[['순번','개봉연도','영화명','감독','등급','국적','장르','전국매출액','전국관객수','전국스크린수','관람객평균평점','전문가평점평균','왓챠평균평점']]
```
